## Comparação do preço com as qualitativas

```r
ggplot(diamantes, aes(x=color, y=price)) + geom_boxplot() + labs(x = "Cor", y = "Preço", caption = "Fonte: Elaborado pelo autor")
```

```r
j <- diamantes[diamantes$color == "J" ,]
dia <- j[j$price >= quantile(j$price,0.75) + 1.5 * (quantile(j$price,0.75) - quantile(j$price,0.25)) ,]
nrow(dia)
nrow(dia) / nrow(diamantes) * 100
summary(dia$price)
```

```r
d <- diamantes[diamantes$color == "D" ,]
dia <- d[d$price >= quantile(d$price,0.75) + 1.5 * (quantile(d$price,0.75) - quantile(d$price,0.25)) ,]
nrow(dia)
nrow(dia) / nrow(diamantes) * 100
summary(dia$price)
```

```r
# H0: Média de J <= Média de D  /  H1: Média de J > Média de D
# Teste Unialdal superior - Rejeitar se z >= z_alpha

media_d <- mean(d$price)
desv_pad_d <- sd(d$price)
n_d <- nrow(d)

media_j <- mean(j$price)
desv_pad_j <- sd(j$price)
n_j <- nrow(j)

significancia <- 0.05
confianca <- 1 - significancia
d_0 <- 0

z <- (media_j - media_d - d_0) / sqrt((desv_pad_j ** 2 / n_j) + (desv_pad_d ** 2 / n_d))
z

p_value <- pnorm(z, lower.tail = F)
p_value

p_value <= significancia
```

```r
# H0: Média de J = Média de D  /  H1: Média de J > Média de D
# Teste Unialdal superior - Rejeitar se p_value <= significância

media_d <- mean(d$price)
desv_pad_d <- sd(d$price)
n_d <- nrow(d)

media_j <- mean(j$price)
desv_pad_j <- sd(j$price)
n_j <- nrow(j)

significancia <- 0.05
confianca <- 1 - significancia
d_0 <- 0

z <- (media_j - media_d - d_0) / sqrt((desv_pad_j ** 2 / n_j) + (desv_pad_d ** 2 / n_d))
z

p_value <- pnorm(z, lower.tail = F)
p_value

p_value <= significancia
```

Com o teste acima podemos confirmar com 5% de significância que há diferença entre o preço médio de venda dos diamantes das cores J e D

# Explorando a variável cut

```r
levels(diamantes$cut)
```

```r
ff <- cbind(summary(diamantes$cut))
a <- data.frame(linhas =c(rownames(ff)),
                    colunas = c(ff[1:5]))

  a %>% ggplot(aes(x = linhas, y = colunas)) +
  geom_col() +
  geom_text(aes(label = colunas), 
          vjust = -1) +
  labs(title = "Contagem dos diamantes de cada corte", x="Qualidade do Corte", y="Frequencia") +
  theme_grey(base_size = 10)
```

```r
rel <- cbind(round(summary(diamantes$cut)/nrow(diamantes)* 100,2))
rel 
```

```r
ggplot(diamantes, aes(x=cut, y=price)) + geom_boxplot() + labs(x = "Corte", y = "Preço", caption = "Fonte: Elaborado pelo autor")
```

Com o boxplot acima verificamos que apesar do corte "Ideal" ser considerado o corte de melhor classificação ele possui o menor preço medio, sendo superado por todas as demais categorias. O corte premium, por sua vez, apresenta média e mediana maior do que os demais.

```r
j <- diamantes[diamantes$cut == "Fair" ,]
dia <- j[j$price >= quantile(j$price,0.75) + 1.5 * (quantile(j$price,0.75) - quantile(j$price,0.25)) ,]
mantes <- j[j$price <= quantile(j$price,0.25) - 1.5 * (quantile(j$price,0.75) - quantile(j$price,0.25)) ,]
nrow(dia)
nrow(mantes)
nrow(dia) / nrow(j) * 100
summary(dia$price)
quantile(j$price,0.75) + 1.5 * (quantile(j$price,0.75) - quantile(j$price,0.25)) 
```

```r
d <- diamantes[diamantes$cut == "Premium" ,]
dia <- d[d$price >= quantile(d$price,0.75) + 1.5 * (quantile(d$price,0.75) - quantile(d$price,0.25)) ,]
mantes <-d[d$price <= quantile(d$price,0.25) - 1.5 * (quantile(d$price,0.75) - quantile(d$price,0.25)) ,]
nrow(dia)
nrow(mantes)
nrow(dia) / nrow(d) * 100
summary(dia$price)
quantile(d$price,0.75) + 1.5 * (quantile(d$price,0.75) - quantile(d$price,0.25))
```

As categorias Very Good e Premium visualmente parecem ter um preço médio parecido. Testaremos em seguida se a média de preço pode ser considerada a mesma para as duas categorias.

```r
# H0: Média de Premium == Média de Fair  /  H1: Média de Premium  Média de Fair
# Teste Unialdal superior - Rejeitar se z >= z_alpha

media_d <- median(d$price)
desv_pad_d <- sd(d$price)
n_d <- nrow(d)

media_j <- median(j$price)
desv_pad_j <- sd(j$price)
n_j <- nrow(j)

significancia <- 0.05
confianca <- 1 - significancia
d_0 <- 0

z <- (media_j - media_d - d_0) / sqrt((desv_pad_j ** 2 / n_j) + (desv_pad_d ** 2 / n_d))
z

p_value <- pnorm(z, lower.tail = T)
p_value

p_value <= significancia
```

Com o teste acima podemos confirmar com 5% de significância que há diferença entre o preço médio de venda dos diamantes com corte Premum e Very Good

```r
summary(diamantes$vol)
```

```r
summary(dia$vol)
```

```r

ggplot(data = dia, aes(x = vol, y = price)) + 
    geom_point(size = 1.5, stroke = 0) + 
    geom_smooth(method = lm) + 
    geom_hline(yintercept = mean(dia$price), color = 'black') + 
    geom_vline(xintercept = mean(dia$vol), color = 'black') +
    xlab("Volume em cm3 ") + 
    ylab("Preço em USS") + 
    ggtitle('Relação entre o volume do diamante em cm3 e o Preço') 
```

```r
ggplot(data = dia, aes(x = vol, y= price)) +
geom_point(aes(x = vol, y= price, color = cut))+
geom_smooth(method = lm, size= 2)+
geom_hline(yintercept = mean(dia$price), color = 'red', size= 0.5)+
geom_vline(xintercept= mean(dia$vol), color= 'red', size= 0.5)+
labs(x= 'Volume', y= 'Preço', title = 'Preço x Volume', caption= 'Fonte: Elaborado pelo autor', subtitle = 'Analisando o preço em função do volume')+
theme_classic()+
theme(
plot.title = element_text(size = 17, hjust= 0),
plot.subtitle = element_text(size = 14, hjust= 0),
plot.caption = element_text(size = 14, hjust= 1),
axis.title.y= element_text(size = 15, hjust= 0),
axis.title.x= element_text(size = 15, hjust= 0),
axis.text.x= element_text(size= 13),
axis.text.y= element_text(size= 13),
)
```
